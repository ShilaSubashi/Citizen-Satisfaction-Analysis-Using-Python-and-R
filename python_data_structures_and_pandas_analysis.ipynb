**Question 1:** Data Structures & Foundations

```python
import statistics

neighbourhood_scores = {
    'Blloku': [8, 7, 9, 6, 8, 7],
    'Kombinat': [5, 4, 6, 3, 5, 4],
    'Sauk': [9, 8, 8, 9, 7, 8],
    'Laprakë': [6, 7, 5, 6, 7, 5],
    'Selitë': [7, 8, 7, 6, 8, 7],
    'Kinostudio': [4, 5, 3, 4, 6, 5],
    'Fresku': [8, 9, 7, 8, 8, 9],
    'Don Bosko': [5, 6, 5, 4, 6, 5]
}
```

**(a) Using a dictionary comprehension, create a new dictionary avg_scores that maps each
neighbourhood to its average satisfaction score, rounded to 2 decimal places.**

```python
avg_scores = {}
for name, scores in neighbourhood_scores.items():
    avg = round(sum(scores) / len(scores), 2)
    avg_scores[name] = avg

print("Average satisfaction scores per neighbourhood:")
print(avg_scores)
```

This result shows that Sauk and Fresku have the highest satisfaction levels, while Kombinat and Kinostudio have the lowest.

**(b) Using a list comprehension, create a sorted list of tuples (neighbourhood, average) for
neighbourhoods with an average score above 7.0, sorted in descending order.**

```python
high_avg_areas = [
    (name, avg) for name, avg in avg_scores.items() if avg > 7.0
]
# Sort in descending order
high_avg_areas.sort(key=lambda x: x[1], reverse=True)

print("\nNeighbourhoods with average > 7.0:")
print(high_avg_areas)
```

**(c) Using set operations, find neighbourhoods that appear in BOTH the “high satisfaction” group (average > 6.5) AND the “consistent” group (standard deviation < 1.5).**

Two sets were created:
I defined high_satisfaction as the set of neighbourhoods with an average score above 6.5, and consistent as the set with a standard deviation below 1.5. A set intersection was then used to find neighbourhoods that belong to both groups.The intersection identifies neighbourhoods that are both performing well on average and have low score variability, indicating reliable performance. I used Python’s set comprehension to derive each group and the & operator for intersection. I used statistics.stdev to compute the sample standard deviation.

```python
high_satisfaction = {n for n, avg in avg_scores.items() if avg > 6.5}
consistent = {n for n, s in neighbourhood_scores.items() if statistics.stdev(s) < 1.5}

both_groups = high_satisfaction.intersection(consistent)

print("\nNeighbourhoods both high satisfaction and consistent:")
print(both_groups)
```

The result shows that Blloku, Sauk, Selitë and Fresku not only perform well on average, but their scores are also relatively stable, indicating more consistent citizen experiences

**(d) Create a tuple containing the overall minimum and maximum average scores. Explain in a comment
why a tuple is appropriate here rather than a list.**

```python
min_max_avg = (min(avg_scores.values()), max(avg_scores.values()))
print("\nOverall (min, max) average scores:", min_max_avg)
```

A tuple is appropriate here because:
A tuple expresses an immutable, fixed pair of related values (the two endpoints of the range) that should not be altered. It communicates a read-only property, i.e., the min and max of the data, which is a summary statistic pair.
Using a list would imply you might append or mutate the pair later, which is inappropriate for a static summary.

**Conclusion of Task 1.1**
The analysis suggests that Sauk and Fresku are the highest-performing neighbourhoods overall. Blloku and Selitë also show strong satisfaction. In contrast, Kombinat and Kinostudio appear to have lower citizen satisfaction, which may indicate areas where the municipality should investigate service improvement.

**Task 1.2: Air Quality Data Parsing**

```python
raw_readings = [
    'sensor_01,PM2.5,45.2,2024-01-15',
    'sensor_02,PM10,88.7,2024-01-15',
    'sensor_01,NO2,32.1,2024-01-15',
    'sensor_03,PM2.5,52.8,2024-01-15',
    'sensor_02,PM2.5,41.0,2024-01-15',
    'sensor_01,PM10,INVALID,2024-01-15',
    'sensor_03,NO2,28.5,2024-01-15'
]
```

**(a) Write a function parse_readings(raw_data) that processes each string and returns a list of dictionaries
with keys ‘sensor’, ‘pollutant’, ‘value’, and ‘date’. Use try-except to handle invalid numeric values — skip
records where the value cannot be converted to float and print a warning message.**
The function parse_readings(raw_data) processes and converts each raw string by splitting it into a list of usable sensor readings: sensor, pollutant, value, and date. If conversion fails, the invalid record is skipped and a warning is printed. This ensures that only clean and usable data is included in later analysis.

```python
def parse_readings(raw_data):
    parsed = []
    for record in raw_data:
        sensor, pollutant, value, date = record.split(',')
        try:
            value = float(value)
            parsed.append({
                'sensor': sensor,
                'pollutant': pollutant,
                'value': value,
                'date': date
            })
        except ValueError:
            print(f"⚠️ Skipping invalid reading: {record}")
    return parsed

parsed_readings = parse_readings(raw_readings)

print("\nParsed readings:")
for r in parsed_readings:
    print(r)
```

**(b) From the parsed data, use a dictionary comprehension to create a dictionary mapping each sensor_id
to the count of valid readings it produced.**

```python
sensor_counts = {}
for record in parsed_readings:
    sensor = record['sensor']
    sensor_counts[sensor] = sensor_counts.get(sensor, 0) + 1

print("\nValid reading counts per sensor:")
print(sensor_counts)
```

**(c) Using list comprehension with conditional filtering, extract all PM2.5 readings and compute their average. Display which sensors reported PM2.5 values above this average.**

```python
pm25_values = [r['value'] for r in parsed_readings if r['pollutant'] == 'PM2.5']
pm25_avg = sum(pm25_values) / len(pm25_values)

above_avg_sensors = [r['sensor'] for r in parsed_readings
                     if r['pollutant'] == 'PM2.5' and r['value'] > pm25_avg]

print("\nAverage PM2.5 value:", round(pm25_avg, 2))
print("Sensors with PM2.5 above average:", above_avg_sensors)
```

**(d) Create a set of all unique pollutant types recorded and a set of all unique sensors. Display both sets.**

```python
unique_pollutants = {r['pollutant'] for r in parsed_readings}
unique_sensors = {r['sensor'] for r in parsed_readings}

print("\nPollutants recorded:", unique_pollutants)
print("Sensors reporting data:", unique_sensors)
```

Sets are useful here because they automatically remove duplicates and clearly show the distinct categories prepes and readings from three separate sensors

Task 2.1: Citizen Survey DataFrame

```python
import pandas as pd
import numpy as np

survey = pd.DataFrame({
    'respondent_id': range(1, 17),
    'neighbourhood': ['Blloku', 'Blloku', 'Kombinat', 'Kombinat',
                      'Sauk','Sauk','Laprakë','Laprakë',
                      'Selitë','Selitë','Kinostudio','Kinostudio',
                      'Fresku','Fresku','Don Bosko','Don Bosko'],
    'age_group': ['18-30','31-45','18-30','46-60',
                  '31-45','18-30','46-60','31-45',
                  '18-30','46-60','31-45','18-30',
                  '46-60','31-45','18-30','46-60'],
    'transport_score': [8,7,5,4,9,8,6,7,7,6,4,5,8,9,5,4],
    'safety_score': [7,8,4,5,8,9,5,6,8,7,3,4,9,8,6,5],
    'green_space_score': [9,8,3,4,8,7,5,6,6,7,5,4,7,8,4,3],
    'overall_satisfaction': [8.0,7.5,4.0,4.5,8.5,8.0,5.5,6.5,
                             7.0,6.5,4.0,4.5,8.0,8.5,5.0,4.0]
})
```

**(a) Display the first 5 rows, the shape, data types, and basic descriptive statistics using .describe().**

```python
print("First 5 rows:")
print(survey.head())

print("\nShape:")
print(survey.shape)

print("\nData types:")
print(survey.dtypes)

print("\nDescriptive statistics:")
print(survey.describe())
```

**(b) Add a new column ‘avg_service_score’ that computes the average of transport_score, safety_score,
and green_space_score for each respondent.**

```python
survey['avg_service_score'] = survey[
    ['transport_score', 'safety_score', 'green_space_score']
].mean(axis=1)

print("\nDataFrame with avg_service_score:")
print(survey.head())
```

**(c) Group the data by ‘neighbourhood’ and compute the mean of all numeric columns. Sort the result by
‘overall_satisfaction’ in descending order. Which neighbourhood ranks highest?**

```python
neighbourhood_means = survey.groupby('neighbourhood').mean(numeric_only=True)
neighbourhood_means = neighbourhood_means.sort_values(
    by='overall_satisfaction',
    ascending=False
)

print("\nMean scores by neighbourhood:")
print(neighbourhood_means)

print("\nHighest ranked neighbourhood:")
print(neighbourhood_means.index[0])
```

 **(d)Filter the DataFrame to show only respondents with overall_satisfaction >= 7.0 AND belonging to the to the ‘18-30’ or ‘31-45’ age groups. How many respondents meet these criteria?**

```python
filtered_respondents = survey[
    (survey['overall_satisfaction'] >= 7.0) &
    (survey['age_group'].isin(['18-30', '31-45']))
]

print("\nFiltered respondents:")
print(filtered_respondents)

print("\nNumber of respondents meeting criteria:")
print(filtered_respondents.shape[0])
```

**(e) Create a new categorical column ‘satisfaction_level’ that classifies respondents as: ‘Low’ (< 5.0),
‘Medium’ (5.0–7.0), or ‘High’ (> 7.0). Display the value_counts() of this new column.**

```python
# Adjust to include exactly 5.0 and 7.0 in Medium
survey['satisfaction_level'] = np.select(
    [
        survey['overall_satisfaction'] < 5.0,
        (survey['overall_satisfaction'] >= 5.0) & (survey['overall_satisfaction'] <= 7.0),
        survey['overall_satisfaction'] > 7.0
    ],
    ['Low', 'Medium', 'High'],
    default='Unknown' # Added a default string to avoid TypeError
)

print("\nSatisfaction level counts:")
print(survey['satisfaction_level'].value_counts())
```

This shows that:
6 respondents are highly satisfied,
5 respondents are moderately satisfied,
5 respondents are in the low satisfaction category.

**Task 2.2: Data Merging & Insights**

```python
profiles = pd.DataFrame({
    'neighbourhood': ['Blloku','Kombinat','Sauk','Laprakë',
                      'Selitë','Kinostudio','Fresku','Don Bosko'],
    'population': [15000,22000,12000,25000,18000,20000,10000,16000],
    'area_km2': [1.8, 3.5, 2.2, 4.0, 2.8, 3.0, 1.5, 2.5],
    'avg_income_eur': [1200, 650, 1400, 580, 900, 700, 1100, 750]
})

# Calculate neighbourhood_means (moved from cell XZP_1R3qUnkD)
neighbourhood_means = survey.groupby('neighbourhood').mean(numeric_only=True)
neighbourhood_means = neighbourhood_means.sort_values(
    by='overall_satisfaction',
    ascending=False
)

# Reset index so neighbourhood becomes a column
neighbourhood_means_reset = neighbourhood_means.reset_index()
```

**(a) Merge the average neighbourhood satisfaction scores (from Task 2.1(c)) with the profiles DataFrame
using an inner join on ‘neighbourhood’. Display the merged result.**

```python
merged = pd.merge(
    neighbourhood_means_reset,
    profiles,
    on='neighbourhood',
    how='inner'
)

print("Merged DataFrame:")
print(merged)
```

**(b) In the merged DataFrame, compute ‘population_density’ (population / area_km2). Which
neighbourhood has the highest population density?**

```python
merged['population_density'] = merged['population'] / merged['area_km2']

highest_density = merged.loc[merged['population_density'].idxmax(), 'neighbourhood']

print("\nPopulation density:")
print(merged[['neighbourhood', 'population_density']])

print("\nHighest population density:")
print(highest_density)
```

**(c) Compute ‘satisfaction_per_income’ (overall_satisfaction / avg_income_eur * 1000). Which
neighbourhood has the best satisfaction relative to income?**

```python
merged['satisfaction_per_income'] = (
    merged['overall_satisfaction'] / merged['avg_income_eur'] * 1000
)

best_relative_satisfaction = merged.loc[
    merged['satisfaction_per_income'].idxmax(),
    'neighbourhood'
]

print("\nSatisfaction per income:")
print(merged[['neighbourhood', 'satisfaction_per_income']])

print("\nBest satisfaction relative to income:")
print(best_relative_satisfaction)
```

**(d) Write the final merged DataFrame to a CSV file named ‘neighbourhood_analysis.csv’. Include code to
read it back and verify the shape matches the original.**

```python
merged.to_csv('neighbourhood_analysis.csv', index=False)

loaded = pd.read_csv('neighbourhood_analysis.csv')

print("\nOriginal merged shape:", merged.shape)
print("Loaded CSV shape:", loaded.shape)
print("Shapes match:", merged.shape == loaded.shape)
```

**Conclusion**:
The analysis shows that Fresku and Sauk have the highest overall satisfaction, Blloku has the highest population density, and Laprakë performs best when satisfaction is considered relative to income.
